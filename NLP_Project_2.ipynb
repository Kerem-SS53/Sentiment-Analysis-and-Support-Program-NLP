```python
from textblob import TextBlob

def analyze_sentiment(sentence):
    blob = TextBlob(sentence)
    sentiment = blob.sentiment.polarity
    return sentiment

def provide_advice(sentiment):
    if sentiment > 0.5:
        return "You seem to be feeling positive. Don't let your mood get down!"
    elif sentiment < -0.5:
        return "You seemed negative today. It might be good for you to share what's bothering you with a friend or professional."
    else:
        return "Your sentiment seems neutral. Go straight!"

def main():
    print("Welcome to the NLP-Psychological Support Program!")
    input_sentence = input("How do you feel yourself today? ")

    sentiment = analyze_sentiment(input_sentence)
    advice = provide_advice(sentiment)

    print("\n Your Mood and My Advice:")
    print(advice)

if __name__ == "__main__":
    main()
```
